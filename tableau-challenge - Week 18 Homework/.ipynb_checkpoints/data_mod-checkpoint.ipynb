```python
# Importing necessary libaries
import pandas as pd
from pathlib import Path
from geopy import distance

# Creating file paths
october = Path('raw_data/JC-202309-citibike-tripdata.csv')
november = Path('raw_data/JC-202310-citibike-tripdata.csv')
december = Path('raw_data/JC-202311-citibike-tripdata.csv')

# Creating DataFrames from .csv files
sept_df = pd.read_csv(october)
oct_df = pd.read_csv(november)
nov_df = pd.read_csv(december)
```

**Creating ride dataset**

```python
# Displaying DataFrames

sept_df.head(5)
```

```python
oct_df.head(5)
```

```python
nov_df.head(5)
```

```python
# Modifications to DataFrames before Union

sept_df['month'] = 'September'
oct_df['month'] = 'October'
nov_df['month'] = 'November'
```

```python
# Checking if modifications are done

oct_df.head(5)
```

```python
# Union
full_df = pd.concat([sept_df, oct_df, nov_df])

# Removing any null values
full_df.dropna()

# Resetting index
full_df.reset_index(drop=True, inplace=True)

full_df
```

```python
# Since the start point and end point are provided, displacement can be measured.

# Create new column
full_df['displacement'] = 0

# Iteration to find displacement for every row
for i in range(len(full_df)):
    try:
        start_coord = (full_df.iloc[i]['start_lat'], full_df.iloc[i]['start_lng'])
        end_coord = (full_df.iloc[i]['end_lat'], full_df.iloc[i]['end_lng'])

        disp = distance.distance(start_coord, end_coord).km
        full_df.loc[i, 'displacement'] = disp
    except ValueError:
        full_df.loc[i, 'displacement'] = 0
```

```python
full_df.head(5)
```

```python
# Cleaning DataFrame to only include columns used for Tableau
clean_df = full_df[['ride_id', 'rideable_type', 'member_casual', 'started_at', 'month', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'displacement']]
clean_df.head(5)
```

```python
# Creating new csv file
clean_df.to_csv('clean_data.csv', index=False)
```

**Creating coordinate dataset**

```python
# Iteration to collect unique station names and their coordinates.

# Preparing DataFrame and empty lists
station_col = full_df[['start_station_name', 'start_lat', 'start_lng']]
stations = []
lat = []
lng = []

# First, determine unique stations
for i in range(len(station_col)):
    if station_col.iloc[i]['start_station_name'] not in stations:
        stations.append(station_col.iloc[i]['start_station_name'])
        
# Then for each station, determine coordinates by getting the average of the latitude and longitude to reduce outlier impact
for i in range(len(stations)):
    station = stations[i]
    station_coords = station_col[station_col['start_station_name'] == station]
    lat.append(station_coords['start_lat'].mean())
    lng.append(station_coords['start_lng'].mean())
```

```python
# Checking whether all 
print(len(stations))
print(len(lat))
print(len(lng))
```

```python
# Creating DataFrame of stations and coords
station_df = pd.DataFrame({'start_station_name': stations, 'lat': lat, 'lng': lng})

station_df.head(5)
```

```python
# Creating new csv file
station_df.to_csv('stations.csv', index=False)
```
