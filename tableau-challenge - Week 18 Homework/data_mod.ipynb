```python
# Importing necessary libaries
import pandas as pd
from pathlib import Path
from geopy import distance

# Creating file paths
december = Path('raw_data/JC-202212-citibike-tripdata.csv')
january = Path('raw_data/JC-202301-citibike-tripdata.csv')
february = Path('raw_data/JC-202302-citibike-tripdata.csv')
march = Path('raw_data/JC-202303-citibike-tripdata.csv')
april = Path('raw_data/JC-202304-citibike-tripdata.csv')
may = Path('raw_data/JC-202305-citibike-tripdata.csv')
june = Path('raw_data/JC-202306-citibike-tripdata.csv')
july = Path('raw_data/JC-202307-citibike-tripdata.csv')
august = Path('raw_data/JC-202308-citibike-tripdata.csv')
september = Path('raw_data/JC-202309-citibike-tripdata.csv')
october = Path('raw_data/JC-202310-citibike-tripdata.csv')
november = Path('raw_data/JC-202311-citibike-tripdata.csv')

# Creating DataFrames from .csv files
dec_df = pd.read_csv(december)
jan_df = pd.read_csv(january)
feb_df = pd.read_csv(february)
mar_df = pd.read_csv(march)
apr_df = pd.read_csv(april)
may_df = pd.read_csv(may)
jun_df = pd.read_csv(june)
jul_df = pd.read_csv(july)
aug_df = pd.read_csv(august)
sept_df = pd.read_csv(september)
oct_df = pd.read_csv(october)
nov_df = pd.read_csv(november)
```

**Creating ride dataset**

```python
# Displaying DataFrames

sept_df.head(5)
```

```python
# Modifications to DataFrames before Union

dec_df['month-year'] = 'December 2022'
jan_df['month-year'] = 'January 2023'
feb_df['month-year'] = 'February 2023'
mar_df['month-year'] = 'March 2023'
apr_df['month-year'] = 'April 2023'
may_df['month-year'] = 'May 2023'
jun_df['month-year'] = 'June 2023'
jul_df['month-year'] = 'July 2023'
aug_df['month-year'] = 'August 2023'
sept_df['month-year'] = 'September 2023'
oct_df['month-year'] = 'October 2023'
nov_df['month-year'] = 'November 2023'
```

```python
# Checking if modifications are done

oct_df.head(5)
```

```python
# Union
full_df = pd.concat([dec_df, jan_df, feb_df, mar_df, apr_df, may_df, jun_df, jul_df, aug_df, sept_df, oct_df, nov_df])

# Removing any null values
full_df.dropna()

# Resetting index
full_df.reset_index(drop=True, inplace=True)

full_df
```

```python
# Since the start point and end point are provided, displacement can be measured.

# Create new column
full_df['displacement'] = 0

# Iteration to find displacement for every row
for i in range(len(full_df)):
    try:
        start_coord = (full_df.iloc[i]['start_lat'], full_df.iloc[i]['start_lng'])
        end_coord = (full_df.iloc[i]['end_lat'], full_df.iloc[i]['end_lng'])

        disp = distance.distance(start_coord, end_coord).km
        full_df.loc[i, 'displacement'] = disp
    except ValueError:
        full_df.loc[i, 'displacement'] = 0
```

```python
# Cleaning DataFrame to only include columns used for Tableau
clean_df = full_df[['ride_id', 'rideable_type', 'member_casual', 'started_at', 'month-year', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'displacement']]
clean_df.head(5)
```

```python
# Creating new csv file
clean_df.to_csv('clean_data_yearround.csv', index=False)
```
