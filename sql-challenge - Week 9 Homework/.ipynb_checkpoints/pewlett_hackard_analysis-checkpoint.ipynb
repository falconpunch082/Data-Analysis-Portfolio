## Data Analysis

```python
# Importing necessary libaries
import pandas as pd
from pathlib import Path

## Preparing data to be displayed
emp_salary_csv = Path("output/emp_salary.csv")
emp_hired_csv = Path("output/emp_hired_1986.csv")
managers_csv = Path("output/managers.csv")
emp_dept_csv = Path("output/emp_dept.csv")
hercules_b_csv = Path("output/hercules_b.csv")
sales_emp_csv = Path("output/sales_emp.csv")
sales_dev_emps_csv = Path("output/sales_dev_emps.csv")
last_name_freq_csv = Path("output/last_name_freq.csv")

emp_salary = pd.read_csv(emp_salary_csv)
emp_hired = pd.read_csv(emp_hired_csv)
managers = pd.read_csv(managers_csv)
emp_dept = pd.read_csv(emp_dept_csv)
hercules_b = pd.read_csv(hercules_b_csv)
sales_emp = pd.read_csv(sales_emp_csv)
sales_dev_emps = pd.read_csv(sales_dev_emps_csv)
last_name_freq = pd.read_csv(last_name_freq_csv)
```

List the employee number, last name, first name, sex, and salary of each employee.

```python
emp_salary = emp_salary.rename(columns={"emp_no": "Employee #", "last_name": "Last Name", "first_name": "First Name",
                                        "sex": "Sex", "salary": "Salary"})

emp_salary
```

List the first name, last name, and hire date for the employees who were hired in 1986.

```python
emp_hired = emp_hired.rename(columns={"last_name": "Last Name", "first_name": "First Name", "hire_date": "Hire Date"})

emp_hired
```

List the manager of each department along with their department number, department name, employee number, last name, and first name.

```python
managers = managers.rename(columns={"dept_no": "Department #", "dept_name": "Department Name", "emp_no": "Employee #",
                                    "first_name": "First Name", "last_name": "Last Name"})

managers
```

List the first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.

```python
hercules_b = hercules_b.rename(columns={"last_name": "Last Name", "first_name": "First Name", "sex": "Sex"})

hercules_b
```

List each employee in the Sales department, including their employee number, last name, and first name.

```python
sales_emp = sales_emp.rename(columns={"last_name": "Last Name", "first_name": "First Name", "emp_no": "Employee #"})

sales_emp
```

List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.

```python
sales_dev_emps = sales_dev_emps.rename(columns={"last_name": "Last Name", "first_name": "First Name", "emp_no": "Employee #",
                                               "dept_name": "Department Name"})

sales_dev_emps
```

List the frequency counts, in descending order, of all the employee last names (that is, how many employees share each last name).

```python
last_name_freq = last_name_freq.rename(columns={"last_name": "Last Name", "frequency": "Frequency"})

last_name_freq
```
